```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)
skip = 0
face_data = []
data_path = 'faceData/'

filename = input("Enter the user name")
cascades = cv2.CascadeClassifier("../cascades/data/haarcascade_frontalface_alt.xml")
while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if ret == False:
        continue
    
    faces = cascades.detectMultiScale(gray_frame, 1.3, 5)
    faces = sorted(faces, key = lambda f:f[2]*f[3], reverse = True)
    for (x, y, w, h) in faces[0:1]:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (233, 12, 123), 2)
        offset = 10
        small_frame = frame[y-offset:y+h+offset, x-offset:x+w+offset]
        face_section = cv2.resize(small_frame, (100, 100))
        
        skip += 1
        if skip%10 == 0:
            face_data.append(face_section)
            print(len(face_data))
            
    
    
    cv2.imshow("frame", frame)
    cv2.imshow("small_frame", small_frame)
    key_pressed = cv2.waitKey(1) & 0xFF
    if(key_pressed == ord('q')):
        break
face_data = np.asarray(face_data)
face_data = face_data.reshape((face_data.shape[0], -1))
print(face_data.shape) 

np.save(data_path+filename+'.npy', face_data)
print("Data successfully stored at : "+data_path+filename+'.npy')
cap.release()
cv2.destroyAllWindows()
```
